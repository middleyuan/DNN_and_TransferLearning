```python
import numpy as np
import os
import math
import tensorflow as tf
from sklearn import preprocessing
from tensorflow.contrib.layers import variance_scaling_initializer
from sklearn import metrics
from sklearn.model_selection import train_test_split

###### Do not modify here ###### 

# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)

reset_graph()

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/")

# training on MNIST but only on digits 0 to 4
X_train1 = mnist.train.images[mnist.train.labels < 5]
y_train1 = mnist.train.labels[mnist.train.labels < 5]
X_valid1 = mnist.validation.images[mnist.validation.labels < 5]
y_valid1 = mnist.validation.labels[mnist.validation.labels < 5]
X_test1 = mnist.test.images[mnist.test.labels < 5]
y_test1 = mnist.test.labels[mnist.test.labels < 5]

###### Do not modify here ###### 

he_init = variance_scaling_initializer()            #Using He initialization
X = tf.placeholder(tf.float32, [None, 784], name='X', )
y = tf.placeholder(tf.int64, [None,], name='y')
keep_prob = tf.placeholder(tf.float32)


layer1 = tf.layers.dense(X, 128, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'layer1')
layer1_drop = tf.layers.dropout(inputs = layer1, rate = keep_prob, name = 'layer1_drop')
layer2 = tf.layers.dense(layer1_drop, 128, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'layer2')
layer2_drop = tf.layers.dropout(inputs = layer2, rate = keep_prob, name = 'layer2_drop')
layer3 = tf.layers.dense(layer2_drop, 128, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'layer3')
layer3_drop = tf.layers.dropout(inputs = layer3, rate = keep_prob, name = 'layer3_drop')
layer4 = tf.layers.dense(layer3_drop, 128, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'layer4')
layer4_drop = tf.layers.dropout(inputs = layer4, rate = keep_prob, name = 'layer4_drop')
layer5 = tf.layers.dense(layer4_drop, 128, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'layer5')
layer5_drop = tf.layers.dropout(inputs = layer5, rate = keep_prob, name = 'layer5_drop')
logits = tf.layers.dense(layer5_drop, 5, activation = tf.nn.elu, kernel_initializer = he_init, bias_initializer = he_init, name = 'logits')

Y_proba = tf.nn.softmax(logits, name = 'Y_proba');
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits = Y_proba, labels = y), name = 'loss')
optimizer = tf.train.AdamOptimizer(0.001).minimize(loss, name = 'training_op')

correct = tf.nn.in_top_k(logits, y, 1)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32), name = 'accuracy')

                                
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
                                
for i in range(200):
    sess.run(optimizer, feed_dict={X: X_train1, y: y_train1, keep_prob: 0.5})
    if i % 10 == 0:
        print(int(i),' Test accuracy:', sess.run(accuracy, feed_dict={X: X_test1, y: y_test1, keep_prob: 1.0}))
```
